```python
import soap
#from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
#rc('text', usetex=True)
from soap import cw
import numpy as np
import matplotlib.pyplot as plt
```

```python
fmin,fmax = 180.2,180.3
tstart,T,tsft = 931035615.,22538,1800.
```

```python
sig = cw.GenerateSignal()
sig.alpha = np.pi
sig.delta = np.arcsin(0.0)
sig.cosi = np.pi
sig.phi0 = np.pi
sig.psi = np.pi/2
sig.f = [180.25,1e-13]
sig.earth_ephem = "/Users/josephbayley/repositories/soap/soap/earth00-19-DE200.dat.gz"
sig.sun_ephem = "/Users/josephbayley/repositories/soap/soap/sun00-19-DE200.dat.gz"
sig.tref = tstart
```

```python
dfmin,dfmax = 180.245,180.255
data = sig.get_spectrogram(tstart = tstart, nsft = 300,tsft=1800,fmin=dfmin,fmax=dfmax,dets= ["H1"],snr=1000,doppler=False,antenna=False)
```

```python
fig, ax =plt.subplots(figsize=(20,9))
fontsize=24
im = ax.imshow(data.H1.norm_sft_power.T,aspect="auto",origin="lower",extent=[min(data.epochs),max(data.epochs),min(data.H1.frequencies),max(data.H1.frequencies)+1./1800],cmap="YlGnBu_r")
ax.plot(data.epochs,data.H1.pulsar_path,"r",label="pulsar frequency")
ax.set_xlabel("Gps time [s]",fontsize=fontsize)
ax.set_ylabel("Frequency [Hz]",fontsize=fontsize)
ax.tick_params(axis='both', which='major', labelsize=17)
ax.legend(fancybox=True,fontsize=fontsize-2)
cbar = fig.colorbar(im,ax=ax)
cbar.ax.tick_params(labelsize=17)
cbar.ax.set_ylabel("FFT power",fontsize=fontsize)
```
